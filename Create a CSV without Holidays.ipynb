```python
accidents = pd.read_csv("datasets/1.-Transportation/accidents-2017.csv")
```

```python
sin_juan = accidents.drop(accidents[(accidents["Month"] == "June") & (accidents["Day"].isin([23, 24]))].index)
```

```python
sin__juan_merce = sin_juan.drop(sin_juan[(sin_juan["Month"] == "September") & (sin_juan["Day"].isin([22, 23, 24, 25]))].index)
```

```python
sin_juan_merce_xmas = sin_juan_merce.drop(sin_juan_merce[(sin_juan_merce["Month"] == "December") & (sin_juan_merce["Day"].isin([24, 25]))].index)
```

```python
sin_juan_merce_xmas_dec8 = sin_juan_merce_xmas.drop(sin_juan_merce_xmas[(sin_juan_merce_xmas["Month"] == "December") & (sin_juan_merce_xmas["Day"].isin([6, 7, 8]))].index)
```

```python
sin_juan_merce_xmas_dec8_hweek = sin_juan_merce_xmas_dec8.drop(sin_juan_merce_xmas_dec8[(sin_juan_merce_xmas_dec8["Month"] == "April") & (sin_juan_merce_xmas_dec8["Day"].isin([14, 15, 16, 17]))].index)
```

```python
sin_juan_merce_xmas_dec8_hweek_dayofc = sin_juan_merce_xmas_dec8_hweek.drop(sin_juan_merce_xmas_dec8_hweek[(sin_juan_merce_xmas_dec8_hweek["Month"] == "September") & (sin_juan_merce_xmas_dec8_hweek["Day"].isin([8, 9, 10, 11]))].index)
```

```python
sin_juan_merce_xmas_dec8_hweek_dayofc_nye = sin_juan_merce_xmas_dec8_hweek_dayofc.drop(sin_juan_merce_xmas_dec8_hweek_dayofc[(sin_juan_merce_xmas_dec8_hweek_dayofc["Month"] == "December") & (sin_juan_merce_xmas_dec8_hweek_dayofc["Day"].isin([31]))].index)
```

```python
sin_juan_merce_xmas_dec8_hweek_dayofc_nye_ny = sin_juan_merce_xmas_dec8_hweek_dayofc_nye.drop(sin_juan_merce_xmas_dec8_hweek_dayofc_nye[(sin_juan_merce_xmas_dec8_hweek_dayofc_nye["Month"] == "January") & (sin_juan_merce_xmas_dec8_hweek_dayofc_nye["Day"].isin([1]))].index)
```

```python
sin_juan_merce_xmas_dec8_hweek_dayofc_nye_ny.to_csv("datasets/1.-Transportation/sin_holidays.csv")
```
